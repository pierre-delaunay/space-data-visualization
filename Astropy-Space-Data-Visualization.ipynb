## Imports

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from astropy.visualization import make_lupton_rgb
from astropy import units as u
```

```python
# Matplotlib parameter
plt.rcParams["figure.figsize"] = (15, 15)
```

## Spiral Galaxy Messier 83 (M83)

### Data analysis

```python
# From ESO Online Digitized Sky Survey (http://archive.eso.org/dss/dss)
M83 = fits.open('data/m83.fits')
M83.info()
```

```python
image_data = M83[0].data

print(type(image_data))
print(image_data.shape)
```

```python
print("Min:", np.min(image_data))
print("Max:", np.max(image_data))
print("Mean:", np.mean(image_data))
print("Stdev:", np.std(image_data))
```

```python
plt.imshow(image_data, cmap='gray')
plt.colorbar()
plt.show()
```

## Keypoints

```python
image = np.array(image_data / 255, dtype=np.uint8)

sift = cv2.SIFT_create()

kp, des = sift.detectAndCompute(image, None)

# Green keypoints
img1 = cv2.drawKeypoints(image, kp, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

print("Number of SIFT descriptors:", len(kp))
print("SIFT dimension:", len(des[0]))

plt.plot(1, 2, 1), plt.imshow(img1), plt.title('SIFT')
plt.show()
```
